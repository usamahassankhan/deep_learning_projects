```python
import tensorflow as tf
mnist=tf.keras.datasets.fashion_mnist
(training_images,training_labels),(test_images,test_labels)=mnist.load_data()
```

```python
training_images
```

```python
test_images#normalize
```

```python
training_images=training_images/255.0
training_images
```

```python
test_images=test_images/255.0
```

```python
print(training_images.size)
print(training_images.ndim)
print(training_images.shape)
print(test_images.shape)
print(test_images.ndim)
print(training_labels.size)
print(training_labels.ndim)
print(training_labels.shape)
print(test_labels.shape)
print(test_labels.ndim)
```

```python
test_images
```

```python
#we use any activation function in  middle we use mostly relu in middel  but in last layer  activation matters
model=tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                tf.keras.layers.Dense(512,activation=tf.nn.relu),
                                tf.keras.layers.Dense(10,activation=tf.nn.softmax)])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(training_images,training_labels,epochs=10,batch_size=128)
model.evaluate(test_images,test_labels)#evaluate tells accuracy of ur model
```

```python
#prediction
a=training_labels[9]
print(a)
import matplotlib.pyplot as plt
plt.imshow(test_images[1])
model.predict_classes(test_images[1].reshape(1, 28, 28 ))
```
